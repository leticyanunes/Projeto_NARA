```python
import pandas as pd
```

```python
produto = pd.read_csv('produtos.csv')
produto
```

```python
produto_novo = pd.read_csv('produtos_sem_marca_corrigidos.csv')
produto_novo
```

```python
novo = pd.merge(produto, produto_novo, left_on='ID_Produto', right_on='ID_Produto',
how='inner')
```

```python
novo = novo.drop(columns = 'Marca_x')
novo
```

```python
adicional = pd.read_csv('Produto adicional.csv', sep = ';')
adicional
```

```python
novo_produto = novo.rename(columns = {'Marca_y':'Marca'})
novo_produto
```

```python
novo_p = pd.concat([novo_produto, adicional], ignore_index = True)
novo_p
```

```python
novo_p.to_csv('Produtos_atualizado', index = False)
```
