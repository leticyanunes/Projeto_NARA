```python
df_vendas = pd.concat([vendas, vendas1, vendas2, vendas3], ignore_index=True)
df_vendas
```

```python
df_vendas['Data'] = df_vendas['Data'].astype(str).str.replace(' 00:00:00', '', regex=False)
df_vendas
```

```python
clientes.drop_duplicates(inplace=True)
produtos.drop_duplicates(inplace=True)
df_vendas.dropna(inplace=True)
avaliacoes.dropna(inplace=True)
```

```python
clientes['Faixa_Etaria'] = np.where(clientes['Idade'] < 26, 'Jovem',
                                    np.where(clientes['Idade'] > 65, 'Idoso', 'Adulto'))

tabela_final = df_vendas.merge(produtos, on='ID_Produto', how='left')
tabela_final = tabela_final.merge(clientes, on='ID_Cliente', how='left')


tabela_final["Valor_Total"] = tabela_final["Quantidade"] * tabela_final["Preco"]

tabela_final["Faixa_Etaria"] = np.where(tabela_final["Idade"] < 26, "Jovem",
                                       np.where(tabela_final["Idade"] > 65, "Idoso", "Adulto"))


perfil = tabela_final.groupby("Faixa_Etaria")["Valor_Total"].sum().reset_index()


total = perfil["Valor_Total"].sum()
perfil["Percentual"] = (perfil["Valor_Total"] / total * 100).round(2)

print("\n--- Valor Total de Compras por Faixa Etária ---")
print(perfil)


plt.pie(
    perfil["Valor_Total"],
    labels=perfil["Faixa_Etaria"],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Valor Total de Compras por Faixa Etária")
plt.show()
```

```python
produtos['Preco'].describe()
sns.boxplot(x=produtos['Preco'])
plt.title("Dispersão de Preços dos Produtos")
plt.show()
```

```python
precos_array = produtos['Preco'].dropna().values


media = np.mean(precos_array)
mediana = np.median(precos_array)


q1 = np.percentile(precos_array, 25)
q2 = np.percentile(precos_array, 50)
q3 = np.percentile(precos_array, 75)
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr


variancia = np.var(precos_array)
desvio_padrao = np.std(precos_array)
coeficiente_variacao = (desvio_padrao / media) * 100
distancia_variancia = (variancia / (media ** 2)) * 100
distancia = (media - mediana) / mediana


print("\n--- Estatísticas Descritivas dos Preços ---")
print(f"Média: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Distância entre média e mediana: {distancia * 100:.2f}%")
print(f"Primeiro quartil (Q1): {q1:.2f}")
print(f"Segundo quartil (Q2 - Mediana): {q2:.2f}")
print(f"Terceiro quartil (Q3): {q3:.2f}")
print(f"Limite inferior: {limite_inferior:.2f}")
print(f"Limite superior: {limite_superior:.2f}")
print("\nVariância:", round(variancia, 2))
print("Desvio padrão:", round(desvio_padrao, 2))
print("Coeficiente de variação:", round(coeficiente_variacao, 2), "%")
print("Distância da variância em relação à média:", round(distancia_variancia, 2), "%")
```

```python
df_vendas['Mes'] = df_vendas['Data'].astype(str).str[5:7]
vendas_mensais = df_vendas.groupby('Mes')['Valor_Total'].sum()
```

```python
Q1 = produtos['Preco'].quantile(0.25)
Q3 = produtos['Preco'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
```

```python
outliers = produtos[(produtos['Preco'] < limite_inferior) | (produtos['Preco'] > limite_superior)]
print(f"Produtos com preços fora do padrão: {len(outliers)}")
```

```python
vendas_produto = df_vendas.groupby('ID_Produto')['Valor_Total'].sum().reset_index()
base_merged = produtos.merge(vendas_produto, on='ID_Produto', how='left')
base_merged = base_merged.merge(avaliacoes.groupby('ID_Produto')['Nota'].mean().reset_index(),
                                on='ID_Produto', how='left')

sns.scatterplot(data=base_merged, x='Preco', y='Valor_Total', hue='Nota', palette='coolwarm')
plt.title("Correlação: Preço x Vendas x Nota Média")
plt.show()

base_merged[['Preco', 'Valor_Total', 'Nota']].corr()
```

```python
# Regressão Linear Simples (quantidade x lucro)

x = df_vendas[['Quantidade']]
y = df_vendas[['Valor_Total']]

modelo = LinearRegression()
modelo.fit(x,y)

y_pred = modelo.predict(x)

plt.Figure(figsize=(8,5))
plt.scatter(x, y, color = 'green')
plt.plot(x, y_pred, color = 'orange')
plt.xlabel('Quantidade vendida')
plt.ylabel('Valor Totad da Venda (R$)')
plt.title('Regrassão Linear - Quantidade x Lucro')
plt.show()
```
