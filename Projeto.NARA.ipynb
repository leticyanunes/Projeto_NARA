```python
vendas = pd.read_csv('vendas.csv', sep=",")
vendas
```

```python
avaliacoes = pd.read_csv('avaliacoes.csv', sep=",")
avaliacoes
```

```python
campanhas = pd.read_csv('campanhas_corrigido.csv', sep=",")
campanhas
```

```python
atendimentos = pd.read_csv('atendimentos.csv', sep=',')
atendimentos
```

```python
atendimentos['Tempo_Resp'] = atendimentos['Tempo_Resposta'] * 60
atendimentos
```

```python
atendimentos = atendimentos.drop('Tempo_Resposta', axis=1)
atendimentos
```

```python
vendas1 = pd.read_csv('venda 1.csv', sep=",")
vendas1
```

```python
vendas1['Valor_Total'] = vendas1['Valor_Total'].astype(float)/100
vendas1
```

```python
vendas2 = pd.read_csv('venda 2.csv', sep=",")
vendas2
```

```python
vendas2['Valor_Total'] = vendas2['Valor_Total'].astype(float)/100
vendas2
```

```python
vendas3 = pd.read_csv('venda 3.csv', sep=",")
vendas3
```

```python
vendas3['Valor_Total'] = vendas3['Valor_Total'].astype(float)/100
vendas3
```

```python
df_vendas = pd.concat([vendas, vendas1, vendas2, vendas3], ignore_index=True)
df_vendas
```

```python
df_vendas['Data'] = df_vendas['Data'].astype(str).str.replace(' 00:00:00', '', regex=False)
df_vendas
```

```python
for nome, df in [("Clientes", clientes), ("Produtos", produtos), ("Vendas", df_vendas),
                 ("Avaliações", avaliacoes), ("Campanhas", campanhas), ("Atendimentos", atendimentos)]:
    print(f"\n--- {nome} ---")
    print(df.info())
    print(df.isnull().sum())
```

```python
clientes.drop_duplicates(inplace=True)
produtos.drop_duplicates(inplace=True)
df_vendas.dropna(inplace=True)
avaliacoes.dropna(inplace=True)
```

```python
produtos['Preco'].describe()
sns.boxplot(x=produtos['Preco'])
plt.title("Dispersão de Preços dos Produtos")
plt.show()
```

```python
df_vendas['Data'] = pd.to_datetime(df_vendas['Data'])
df_vendas.describe()
```

```python
vendas_mensais = df_vendas.groupby(df_vendas['Data'].dt.to_period('M'))['Valor_Total'].sum()
vendas_mensais.plot(kind='line', marker='o', title="Evolução Mensal das Vendas")
plt.show()
```

```python
Q1 = produtos['Preco'].quantile(0.25)
Q3 = produtos['Preco'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
```

```python
outliers = produtos[(produtos['Preco'] < limite_inferior) | (produtos['Preco'] > limite_superior)]
print(f"Produtos com preços fora do padrão: {len(outliers)}")
```

```python
vendas_produto = df_vendas.groupby('ID_Produto')['Valor_Total'].sum().reset_index()
base_merged = produtos.merge(vendas_produto, on='ID_Produto', how='left')
base_merged = base_merged.merge(avaliacoes.groupby('ID_Produto')['Nota'].mean().reset_index(),
                                on='ID_Produto', how='left')

sns.scatterplot(data=base_merged, x='Preco', y='Valor_Total', hue='Nota', palette='coolwarm')
plt.title("Correlação: Preço x Vendas x Nota Média")
plt.show()

base_merged[['Preco', 'Valor_Total', 'Nota']].corr()
```

```python
sns.scatterplot(x='Tempo_Resp', y='Nota_Satisfacao', data=atendimentos)
plt.title("Tempo de Resposta x Nota de Satisfação")
plt.show()

print(atendimentos[['Tempo_Resp', 'Nota_Satisfacao']].corr())
```

```python
df_vendas['Mes'] = df_vendas['Data'].dt.to_period('M')
canal_mensal = df_vendas.groupby(['Mes', 'Canal'])['Valor_Total'].sum().unstack()
canal_mensal.plot(kind='line', marker='o', title="Vendas Mensais por Canal")
plt.show()
```

```python
df_vendas.to_csv("vendas_tratada.csv", index=False)
produtos.to_csv("produtos_tratados.csv", index=False)
clientes.to_csv("clientes_tratados.csv", index=False)
```

```python
# Regressão Linear Simples (quantidade x lucro)

x = df_vendas[['Quantidade']]
y = df_vendas[['Valor_Total']]

modelo = LinearRegression()
modelo.fit(x,y)

y_pred = modelo.predict(x)

plt.Figure(figsize=(8,5))
plt.scatter(x, y, color = 'green')
plt.plot(x, y_pred, color = 'orange')
plt.xlabel('Quantidade vendida')
plt.ylabel('Valor Totad da Venda (R$)')
plt.title('Regrassão Linear - Quantidade x Lucro')
plt.show()
```
